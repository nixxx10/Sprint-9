## Corrección del Ejercicio 2 — Convertidor de temperaturas

> **Nota para el profesor:** esta es la versión corregida. Las fórmulas y la lógica
> de conversión **no se han modificado**; solo se ha añadido la **validación de entradas**
> que pedía el enunciado.

### ✅ Qué se mantiene igual (sin cambios)
- Las cuatro fórmulas de conversión (C↔F↔K).
- La estructura `if / elif / elif`.
- El uso de un **diccionario** como valor de retorno.
- El `return 'ERROR'` cuando la **unidad** no es válida.

### 🔧 Qué se ha corregido
1. **Entradas no numéricas** → se envuelve `float(input(...))` en un bloque
   `try / except ValueError`. Así, si el usuario escribe `"treinta"` en lugar de `30`,
   el programa ya **no se rompe**: devuelve `'ERROR'`.
2. **Valores incoherentes (cero absoluto)** → se añade un diccionario `minimos`
   con el límite físico inferior de cada unidad (−273.15 °C, −459.67 °F, 0 K) y se
   rechaza cualquier valor por debajo de ese límite, devolviendo `'ERROR'`.
3. **Orden de validación** → primero se valida la unidad, luego que el valor sea
   numérico y, por último, que sea físicamente posible. Solo si todo es correcto se convierte.

```python
def convert_temperatures():
    # ── NUEVO: límite físico inferior (cero absoluto) en cada unidad ──
    minimos = {'celsius': -273.15, 'fahrenheit': -459.67, 'kelvin': 0}

    tipo_t = input('Introduzca el tipo de temperatura (Celsius/Fahrenheit/Kelvin): ').lower()

    # ── NUEVO: validar la unidad ANTES de pedir el valor ──
    if tipo_t not in minimos:
        return 'ERROR'

    # ── NUEVO: try/except para entradas no numéricas (p. ej. "treinta") ──
    try:
        valor = float(input('Introduzca la temperatura: '))
    except ValueError:
        return 'ERROR'

    # ── NUEVO: validar que el valor no esté por debajo del cero absoluto ──
    if valor < minimos[tipo_t]:
        return 'ERROR'

    # ── IGUAL QUE ANTES: las fórmulas de conversión no cambian ──
    if tipo_t == 'celsius':
        f = (valor * 9/5) + 32
        k = valor + 273.15
        return {'Celsius': valor, 'Fahrenheit': f, 'Kelvin': k}

    elif tipo_t == 'fahrenheit':
        c = (valor - 32) * 5/9
        k = c + 273.15
        return {'Fahrenheit': valor, 'Celsius': c, 'Kelvin': k}

    elif tipo_t == 'kelvin':
        c = valor - 273.15
        f = (c * 9/5) + 32
        return {'Kelvin': valor, 'Celsius': c, 'Fahrenheit': f}
```

```python
# Pruebas de la validación (ejecuta y prueba cada caso):
# 1) Unidad válida + número válido     -> celsius / 39      -> devuelve diccionario
# 2) Texto en vez de número            -> celsius / treinta -> devuelve 'ERROR'
# 3) Valor por debajo del cero absoluto -> celsius / -300    -> devuelve 'ERROR'
# 4) Unidad inexistente                -> grados  / 20      -> devuelve 'ERROR'
resultado = convert_temperatures()
print(resultado)
```
