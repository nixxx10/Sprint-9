# 1. Calculadora del Ă­ndice de masa corporal (IMC)

Para calcular el IMC se divide el peso (kg) entre el cuadrado de la altura (m):

$$IMC = \frac{peso}{altura^2}$$

ClasificaciĂłn segĂşn el resultado:

| Rango           | CategorĂ­a     |
|-----------------|---------------|
| IMC < 18.5      | Bajo peso     |
| 18.5 â 24.9     | Peso normal   |
| 25 â 29.9       | Sobrepeso     |
| âĽ 30            | Obesidad      |

Si el peso o la altura son negativos, la funciĂłn devuelve `ERROR`.

## FunciĂłn `calcular_imc`

Pide los datos al usuario, calcula el IMC y devuelve la categorĂ­a correspondiente.

````python
def calcular_imc():
    peso = float(input('Introduzca su peso (kg): '))
    altura = float(input('Introduzca su altura (m): '))

    if peso <= 0 or altura <= 0:
        return 'ERROR'
    
    imc = peso / (altura ** 2)

    if imc < 18.5:
        return 'Bajo_peso'
    elif 18.5 <= imc <= 24.9: 
        return 'Peso_normal'
    elif 25 <= imc <= 29.9: 
        return 'Sobrepeso'
    else:
        return 'Obesidad'
````

## EjecuciĂłn

````python
resultado = calcular_imc()
print(resultado)
````

# 2. Convertidor de temperaturas

Conversor entre las tres unidades de temperatura mĂĄs comunes:
**Celsius (Â°C)**, **Fahrenheit (Â°F)** y **Kelvin (K)**.

El usuario introduce la unidad de origen y el valor; la funciĂłn devuelve
un **diccionario** con las tres equivalencias, de modo que el resultado
se puede guardar y consultar despuĂŠs (un `print()` no se podrĂ­a reutilizar).

Si la unidad introducida no es vĂĄlida, devuelve `ERROR`.

## FĂłrmulas usadas

| Origen â Destino       | FĂłrmula                  |
|------------------------|--------------------------|
| Celsius â Fahrenheit   | `F = C Âˇ 9/5 + 32`       |
| Celsius â Kelvin       | `K = C + 273.15`         |
| Fahrenheit â Celsius   | `C = (F â 32) Âˇ 5/9`     |
| Kelvin â Celsius       | `C = K â 273.15`         |

## FunciĂłn `convert_temperatures`

Pide la unidad de origen y el valor, y devuelve un diccionario con las
tres equivalencias. El diccionario es ideal como contenedor: cada clave
es la unidad y cada valor su temperatura correspondiente.

````python
def convert_temperatures():

    tipo_t = input('Introduzca el tipo de temperatura (Celsius/Fahrenheit/Kelvin): ').lower()
    valor = float(input('Introduzca la temperatura: '))

    if tipo_t == 'celsius':
        f = (valor * 9/5) + 32
        k = valor + 273.15
        return {'Celsius': valor, 'Fahrenheit': f, 'Kelvin': k}

    elif tipo_t == 'fahrenheit':
        c = (valor - 32) * 5/9
        k = c + 273.15
        return {'Fahrenheit': valor, 'Celsius': c, 'Kelvin': k}

    elif tipo_t == 'kelvin':
        c = valor - 273.15
        f = (c * 9/5) + 32
        return {'Kelvin': valor, 'Celsius': c, 'Fahrenheit': f}

    else: 
        return 'ERROR'
````

## EjecuciĂłn

El resultado se guarda en `resultado` (un diccionario) y luego se imprime.
Al guardarlo en una variable, podrĂ­amos acceder a cada unidad por separado:

â```python
resultado['Celsius']
resultado['Fahrenheit']
resultado['Kelvin']
â```

````python
resultado = convert_temperatures()
print(resultado)
````

## Exercici 3: Comptador de paraules d'un text

Donada una cadena de text, volem saber quantes vegades apareix cada paraula.

**CasuĂ­stiques gestionades:**
- Entrada que no ĂŠs una cadena (`None`, nĂşmeros, llistes...)
- Text buit o amb nomĂŠs espais
- Text sense cap paraula (nomĂŠs nĂşmeros o sĂ­mbols)
- MajĂşscules i minĂşscules (`Hola` i `hola` compten com la mateixa paraula)
- Signes de puntuaciĂł (`.`, `!`, `?`, `,`, `;`...)
- Accents i carĂ cters catalans (Ă , Ă¨, ĂŠ, Ă­, ĂŻ, Ă˛, Ăł, Ăş, Ăź, Ă§)
- Espais mĂşltiples, salts de lĂ­nia i tabuladors

**EstratĂ¨gia:** fem servir `re.findall` per extreure directament les paraules
(seqĂźĂ¨ncies de lletres), i `collections.Counter` per comptar-les.
Separarem la lĂ˛gica en tres funcions amb una sola responsabilitat cadascuna.

````python
import re
from collections import Counter
````

### FunciĂł auxiliar: `extraure_paraules`

Valida l'entrada i extreu totes les paraules com una llista de cadenes en minĂşscules.
Centralitzar aquĂ­ la validaciĂł evita repetir-la a cada funciĂł.

````python
def extraure_paraules(text):
    """
    Valida i extreu totes les paraules d'un text.
    text   : str       - text d'entrada
    Retorna: list[str] - llista de paraules en minĂşscules | None si hi ha error
    """
    if not isinstance(text, str):
        print(f"AdvertĂ¨ncia: l'entrada ha de ser una cadena de text (str), "
              f"perĂ˛ s'ha rebut {type(text).__name__}.")
        return None

    if not text.strip():
        print("AdvertĂ¨ncia: el text estĂ  buit.")
        return None

    paraules = re.findall(r"[a-zA-ZĂ Ă¨ĂŠĂ­ĂŻĂłĂ˛ĂşĂźĂ§ĂąĂĂĂĂĂĂĂĂĂĂĂ]+", text.lower())

    if not paraules:
        print("AdvertĂ¨ncia: no s'ha trobat cap paraula al text "
              "(pot contenir nomĂŠs nĂşmeros o sĂ­mbols).")
        return None

    return paraules
````

### FunciĂł principal: `comptar_paraules`

Utilitza `extraure_paraules` per obtenir la llista de paraules i `Counter`
per comptar-les. Mostra els resultats ordenats de mĂŠs a menys freqĂźent
grĂ cies al mĂ¨tode `.most_common()` de `Counter`.

````python
def comptar_paraules(text):
    """
    Compta les vegades que apareix cada paraula en un text.
    text   : str     - text d'entrada
    Retorna: Counter - diccionari {paraula: freqĂźĂ¨ncia} | None si hi ha error
    """
    paraules = extraure_paraules(text)
    if paraules is None:
        return None

    comptador = Counter(paraules)

    print(f"Total de paraules : {len(paraules)}")
    print(f"Paraules Ăşniques  : {len(comptador)}\n")
    print(f"{'Paraula':<20} FreqĂźĂ¨ncia")
    print("-" * 30)
    for paraula, frequencia in comptador.most_common():
        print(f"  {paraula:<18} {frequencia}")

    return comptador
````

### Extra: `llargada_mitjana_paraules`

Calcula la llargada mitjana de les paraules del text.
Reutilitza `extraure_paraules` per no repetir la validaciĂł ni el procĂŠs de neteja.

Exemple: `"Hola com va?"` â `(4 + 3 + 2) / 3 = 3.0`

````python
def llargada_mitjana_paraules(text):
    """
    Calcula la llargada mitjana de les paraules d'un text.
    text   : str   - text d'entrada
    Retorna: float - llargada mitjana arrodonida a 2 decimals | None si hi ha error
    """
    paraules = extraure_paraules(text)
    if paraules is None:
        return None

    return round(sum(len(p) for p in paraules) / len(paraules), 2)
````

### DemostraciĂł

````python
# Text de prova amb repeticions, majĂşscules i puntuaciĂł
text_prova = "El gat menja peix. El gat ĂŠs gros i el gat dorm molt. GROS ĂŠs el gat!"
comptar_paraules(text_prova)
````

````python
# VerificaciĂł de l'EXTRA
llargada_mitjana_paraules("Hola com va?")  # â 3.0
````

````python
# VerificaciĂł de l'error
comptar_paraules(None)          # entrada no vĂ lida
comptar_paraules("")            # text buit
comptar_paraules("   ")        # text amb espais
comptar_paraules("123 456 !!!") # sense paraules
````

## Exercici 4: Diccionari invers (amb possibilitat de duplicats)

Donada una enquesta emmagatzemada en un diccionari, necessitem invertir-la:
les claus es converteixen en valors i els valors en claus.

**Casos possibles:**
- Valors Ăşnics â inversiĂł directa
- Valors duplicats â les claus s'agrupen en una llista + missatge d'advertĂ¨ncia
- Entrada no vĂ lida (no ĂŠs un dict, estĂ  buit) â advertĂ¨ncia i `None`

**EstratĂ¨gia:** fem servir `Counter` per detectar duplicats i `setdefault`
per agrupar les claus repetides sense necessitat d'`if/elif`.

### Funcions auxiliars

- `validar_diccionari`: comprova que l'entrada ĂŠs un diccionari no buit.
- `detectar_duplicats`: retorna el conjunt de valors que apareixen mĂŠs d'una vegada,
  usant `Counter` sobre els valors del diccionari.

````python
def validar_diccionari(diccionari):
    """
    Valida que l'entrada ĂŠs un diccionari no buit.
    diccionari : any          - valor a validar
    Retorna    : (True, None) si vĂ lid | (False, str) amb missatge si no
    """
    if not isinstance(diccionari, dict):
        return False, (
            f"AdvertĂ¨ncia: l'entrada ha de ser un diccionari (dict), "
            f"perĂ˛ s'ha rebut {type(diccionari).__name__}."
        )
    if not diccionari:
        return False, "AdvertĂ¨ncia: el diccionari estĂ  buit."
    return True, None


def detectar_duplicats(diccionari):
    """
    Detecta quins valors apareixen mĂŠs d'una vegada al diccionari.
    diccionari : dict - diccionari a analitzar
    Retorna    : set  - conjunt de valors duplicats
    """
    comptador_valors = Counter(diccionari.values())
    return {valor for valor, recompte in comptador_valors.items() if recompte > 1}
````

### FunciĂł principal: `reverse_dictionary`

Orquestra les funcions auxiliars: valida, detecta duplicats i construeix
el diccionari invers. Els valors duplicats s'agrupen en una llista grĂ cies
a `setdefault`, que crea la llista automĂ ticament si no existeix encara.

````python
def reverse_dictionary(diccionari):
    """
    Inverteix un diccionari intercanviant claus i valors.
    Si un valor apareix mĂŠs d'una vegada, les claus s'agrupen en una llista.
    diccionari : dict - diccionari a invertir
    Retorna    : dict - diccionari invertit | None si hi ha error
    """
    valid, error = validar_diccionari(diccionari)
    if not valid:
        print(error)
        return None

    duplicats = detectar_duplicats(diccionari)

    if duplicats:
        print(f"AdvertĂ¨ncia: els segĂźents valors estan duplicats: {duplicats}")

    invers = {}
    for clau, valor in diccionari.items():
        if valor in duplicats:
            invers.setdefault(valor, []).append(clau)
        else:
            invers[valor] = clau

    return invers
````

### DemostraciĂł

````python
# Cas 1: tots els valors Ăşnics
reverse_dictionary({'a': 1, 'b': 2, 'c': 3})
````

````python
# Cas 2: valor duplicat â s'agrupa en llista
reverse_dictionary({'x': 'apple', 'y': 'banana', 'z': 'banana'})
````

````python
# Cas 3: validaciĂł d'errors
reverse_dictionary("no soc un diccionari")
reverse_dictionary({})
````
